## Câu 1
a) Tạo mảng A có kích thước 20x5 với các giá trị từ 0 đến 99

```python
import numpy as np

A = np.arange(100).reshape(20, 5)
A
```

b) Tính tổng, giá trị trung bình của hàng thứ 15 của mảng (2đ)

```python
A[14]
print('Tổng của hàng thứ 15 trong mảng là: ', A[14].sum())
print('Giá trị tb của hàng thứ 15 trong mảng là: ', A[14].mean())
```

c) Biến đổi kích thước về mảng 10x10 và tính giá trị trung bình cột thứ 10

```python
A = A.reshape(10, 10)
print(A)
```

```python
# print(A[:, 9]) #[ 9 19 29 39 49 59 69 79 89 99]
print('Giá trị trung bình của cột thứ 10 là: ', np.mean(A[:, 9]))
```
